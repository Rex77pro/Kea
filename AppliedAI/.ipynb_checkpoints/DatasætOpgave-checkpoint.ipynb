```python
H1: Personer med højere uddannelsesniveau har en signifikant højere gennemsnitlig indkomst og formue end personer med lavere uddannelsesniveau.
H2: Personer med fast beskæftigelse (fx fuldtidsansatte) har højere gennemsnitlig indkomst og formue sammenlignet med personer, 
    der er ledige eller arbejder deltid.
```

```python
import numpy as np
import pandas as pd


csv_path = r"C:\\Users\\smaur\\Code\\Kea\\AppliedAI\\wealth_income_dataset.csv"
df = pd.read_csv(csv_path)

print(df.columns.tolist())
```

```python
print(df[['Income','Wealth']].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for income
plt.figure(figsize=(8,4))
sns.histplot(df['Income'], kde=True)
plt.title('Fordeling af Income')
plt.xlabel('Income')
plt.ylabel('Antal Observationer')
plt.show()

# Boxplot af Income opdelt på Education
plt.figure(figsize=(10,6))
sns.boxplot(x='Education', y='Income', data=df)
plt.title('Income fordelt på Education')
plt.xlabel('Education')
plt.ylabel('Income')
plt.show()
```

```python
# Der er en signifikant coralaiton
```

```python
import scipy.stats as stats

groups = [group['Income'].values for name, group in df.groupby('Education')]
anova_result = stats.f_oneway(*groups)
print("ANOVA for Income på tværs af Education:", anova_result)
```

```python
# En p-værdi på 0.0 (praktisk talt 0) betyder, at der er en ekstremt signifikant forskel i gennemsnitlig indkomst mellem de forskellige grupper for Education. 
# Med andre ord kan vi afvise nul-hypotesen om, at der ikke er nogen forskel i indkomst på tværs af uddannelsesniveauer. 
# Dette indikerer, at uddannelsesniveauet har en signifikant effekt på den gennemsnitlige indkomst i dit datasæt.
```

```python
import statsmodels.formula.api as smf

# Sørg for, at Education og Employment_Status behandles som kategoriske variable
df['Education'] = df['Education'].astype('category')
df['Employment_Status'] = df['Employment_Status'].astype('category')

# Lav regressionsmodellen
model = smf.ols('Income ~ C(Education) + C(Employment_Status) + Age', data=df).fit()
print(model.summary())
```

```python

# Modellen viser, at uddannelsesniveau og alder er stærkt forbundet med indkomst, mens beskæftigelsesstatus (som angivet i de givne kategorier) ikke har en signifikant effekt. 
# Modellen forklarer ca. 47 % af variationen i indkomst, hvilket er en anstændig forklaringsgrad for et datasæt af denne art. 
# Det er vigtigt at overveje, at andre ukendte faktorer også kan spille en rolle i at bestemme en persons indkomst.
```
